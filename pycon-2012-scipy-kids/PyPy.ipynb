PyPy is an alternate implementation of Python, in Python

It implements a just-in-time compiler for Python

PyPy is highly compatible with existing python code. It supports ctypes and can run popular python libraries like twisted and django.

It is written in Python and easy to hack at.  Good for exploratory/research stuff

Pluggable garbage collection

It's translator tool chain can be used to generate JIT compilers for other languages
